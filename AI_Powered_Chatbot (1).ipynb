```python
import nltk

nltk.download('punkt')

print("="*50)
print("🤖 AI POWERED CHATBOT")
print("="*50)
print("Type 'exit' to close the chatbot.\n")

responses = {
    "hello": "Hello! How can I help you?",
    "hi": "Hi! Welcome!",
    "how are you": "I'm doing great. Thank you!",
    "what is python": "Python is a popular programming language.",
    "what is ai": "AI stands for Artificial Intelligence.",
    "what is machine learning": "Machine Learning is a branch of AI that allows computers to learn from data.",
    "what is data science": "Data Science is the study of data to gain useful insights.",
    "who created python": "Python was created by Guido van Rossum.",
    "thank you": "You're welcome!",
    "bye": "Goodbye! Have a nice day!"
}

# Create chat history file
file = open("chat_history.txt", "a")

while True:

    user = input("You : ").lower()

    if user == "exit":
        print("🤖 Chatbot : Goodbye!")
        file.write("User: exit\n")
        file.write("Chatbot: Goodbye!\n")
        break

    elif user in responses:
        answer = responses[user]
        print("🤖 Chatbot :", answer)

        file.write("User: " + user + "\n")
        file.write("Chatbot: " + answer + "\n\n")

    else:
        answer = "Sorry, I don't understand that question."
        print("🤖 Chatbot :", answer)

        file.write("User: " + user + "\n")
        file.write("Chatbot: " + answer + "\n\n")

file.close()

print("\n✅ Chat history saved as 'chat_history.txt'")
```
